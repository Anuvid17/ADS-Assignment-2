```python
# ===========================================
# 1. Import Libraries
import pandas as pd

# ===========================================
# 2. Load Dataset
df = pd.read_csv("smartphones.csv")

print("Dataset Loaded ✅")
print("Shape:", df.shape)
df.head()
```

```python
# 3. Explore Dataset

print("Basic Info:")
df.info()

print("\nStatistical Summary:")
display(df.describe(include="all"))

print("\nFirst 10 Rows:")
df.head(10)
```

```python
# 4. Handle Missing Values

print("Missing Values Before:")
print(df.isnull().sum())

# --- Clean & Convert price ---
if "price" in df.columns:
    # Remove ₹ and commas → convert to float
    df['price'] = df['price'].replace('[₹,]', '', regex=True).astype(float)
    # Fill missing prices with median
    df['price'] = df['price'].fillna(df['price'].median())

# --- Handle rating ---
if "rating" in df.columns:
    df['rating'] = df['rating'].fillna(df['rating'].mean())

# --- Handle camera ---
if "camera" in df.columns:
    df['camera'] = df['camera'].fillna(df['camera'].mode()[0])

# --- Handle card (expandable storage info) ---
if "card" in df.columns:
    df['card'] = df['card'].fillna("Not Specified")

# --- Handle OS ---
if "os" in df.columns:
    df['os'] = df['os'].fillna("Unknown")

print("\nMissing Values After:")
print(df.isnull().sum())
```

```python
# 5. Remove Duplicates

print("Duplicates Found:", df.duplicated().sum())
df = df.drop_duplicates()
print("Shape After Removing Duplicates:", df.shape)
```

```python
# 6. Correct Data Types

# Extract numeric values from RAM (e.g., "8 GB" -> 8)
if "ram" in df.columns:
    df['ram_gb'] = df['ram'].str.extract(r'(\d+)').astype(float)

# Extract numeric values from Battery (e.g., "5000 mAh" -> 5000)
if "battery" in df.columns:
    df['battery_mah'] = df['battery'].str.extract(r'(\d+)').astype(float)

# Extract Display size (e.g., "6.5 inches" -> 6.5)
if "display" in df.columns:
    df['display_inches'] = df['display'].str.extract(r'(\d+\.?\d*)').astype(float)

# Convert release_date to datetime (if available)
if "release_date" in df.columns:
    df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

print("Data Types After Cleaning:")
df.info()
```

```python
# 7. Final Cleaned Dataset

print("Preview Cleaned Dataset:")
df.head()

# Save cleaned dataset
df.to_csv("smartphones_cleaned.csv", index=False)
print("✅ Cleaned dataset saved as smartphones_cleaned.csv")
```

```python
# 8. Product Recommendation Search

def recommend_products(keyword, df, column="model"):
    """
    Recommends products containing the keyword (case-insensitive).
    """
    keyword = keyword.lower()
    results = df[df[column].str.lower().str.contains(keyword, na=False)]

    if results.empty:
        print(f"No products found for: {keyword}")
    else:
        return results

# Example Usage:
recommend_products("vivo", df)
```
